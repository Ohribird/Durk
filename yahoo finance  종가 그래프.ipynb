```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import yfinance as yf


ticker = '005930.KS'   # TSLA, 9984.T, 등등 
start = "2020-01-01"
end = "2022-12-31"

df = yf.Ticker(ticker).history(start=start, end= end)
fig,ax = plt.subplots(1,1,figsize=(12,4))   #그림 사이즈 조정, 가로 12인치, 세로 4인치
ax.plot(df.index, df.Close)
plt.show()
```

```python
tickers = ['AAPL','AMZN', 'GOOG']
start = '2021-01-01'
end = '2022-12-31'

fig,ax = plt.subplots(1,1,figsize=(12,4))  
for ticker in tickers:
  df= yf.Ticker(ticker).history(start=start, end=end)
  ax.plot(df.index, df.Close/df.Close[0]*100, label = f'{ticker}')
  #label = f'{ticker}' -> 그래프 우측 상단 
ax.legend(loc= 'best')   #라벨을 어디 붙일지 정하는거 
plt.show()
```
